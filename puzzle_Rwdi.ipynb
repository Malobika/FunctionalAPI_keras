```python
#Fourth Gold Star
import numpy as np
import pandas as pd
#/content/drive/MyDrive/Colab Notebooks/input.rtf
j=[]

with open('/content/drive/MyDrive/Colab Notebooks/4.txt', 'r', encoding='utf-16', errors='ignore') as f:
    lines = f.readlines()

   
k=set()
dict1={"h":0,"aim":0,"depth":0}
for each in lines:
    a,b =each.split(" ")
    #j.append(int(each))
    if a=="forward":
      dict1["h"] += int(b)
      dict1["depth"] += int(b)*dict1["aim"]
    elif a=="down":
      dict1["aim"] += int(b)
    else:
      dict1["aim"] -= int(b)
    
print(dict1)
final =dict1["h"]*dict1["depth"]
print(final)
```
